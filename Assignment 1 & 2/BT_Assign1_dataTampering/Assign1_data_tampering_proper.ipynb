```python
import hashlib
import os

HASH_FILE = "hash_store.txt"

# ---------- Function to calculate hash ----------
def calculate_hash(data):
    return hashlib.sha256(data.encode()).hexdigest()


# ---------- Read data from files ----------
file_names = ["text1.txt", "text2.txt", "text3.txt", "text4.txt", "text5.txt"]

current_hashes = {}

for file_name in file_names:
    with open(file_name, "r") as f:
        data = f.read().strip()
        file_hash = calculate_hash(data)
        current_hashes[file_name] = file_hash


# ---------- First run: store hashes ----------
if not os.path.exists(HASH_FILE):
    print("First run — storing hashes...")

    with open(HASH_FILE, "w") as f:
        for file_name, file_hash in current_hashes.items():
            f.write(f"{file_name}|{file_hash}\n")

    print("Hashes stored successfully.")

else:
    print("Checking for tampering...")

    stored_hashes = {}

    with open(HASH_FILE, "r") as f:
        for line in f:
            file_name, file_hash = line.strip().split("|")
            stored_hashes[file_name] = file_hash

    tampered = False

    for file_name in file_names:
        if current_hashes[file_name] != stored_hashes[file_name]:
            print(f" Data tampered in: {file_name}")
            tampered = True

    if not tampered:
        print(" No tampering detected — files are intact.")
```
