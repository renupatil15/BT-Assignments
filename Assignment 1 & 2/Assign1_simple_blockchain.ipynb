```python
import hashlib
import time
```

```python
class Block:
    def __init__(self, index, data, previous_hash):
        self.index = index
        self.timestamp = time.time()
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        block_string = str(self.index) + str(self.timestamp) + str(self.data) + str(self.previous_hash)
        return hashlib.sha256(block_string.encode()).hexdigest()
```

```python
class Blockchain:
    def __init__(self):
        self.chain = [self.create_a_block()]

    def create_a_block(self):
        return Block(0, "A Node", "0")

    def get_latest_block(self):
        return self.chain[-1]

    def add_block(self, new_data):
        latest_block = self.get_latest_block()
        new_block = Block(
            index=len(self.chain),
            data=new_data,
            previous_hash=latest_block.hash
        )
        self.chain.append(new_block)
```

```python
def main():
    # Create blockchain object
    my_blockchain = Blockchain()

    # Add blocks
    my_blockchain.add_block("Transaction 1: B Node")
    my_blockchain.add_block("Transaction 2: C Node")
    my_blockchain.add_block("Transaction 3: D Node")

    # Display blockchain
    for block in my_blockchain.chain:
        print("Index:", block.index)
        print("Timestamp:", block.timestamp)
        print("Data:", block.data)
        print("Previous Hash:", block.previous_hash)
        print("Hash:", block.hash,"\n")
```

```python
main()
```
