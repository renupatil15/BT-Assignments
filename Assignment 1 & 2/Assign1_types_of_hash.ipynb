```python
import hashlib

# Simple hash function
def get_hash(data):
    return hashlib.sha256(data.encode()).hexdigest()


# Read file data
def read_file(filename):
    file = open(filename, "r")
    data = file.read()
    file.close()
    return data


# 1. Independent Hashing
def independent_hashing():
    print("\nIndependent Hashing")
    d1 = read_file("file1.txt")
    d2 = read_file("file2.txt")
    print(get_hash(d1))
    print(get_hash(d2))


# 2. Repeated Hashing
def repeated_hashing():
    print("\nRepeated Hashing")
    data = read_file("file1.txt")
    h1 = get_hash(data)
    h2 = get_hash(h1)
    print(h2)


# 3. Combined Hashing
def combined_hashing():
    print("\nCombined Hashing")
    d1 = read_file("file1.txt")
    d2 = read_file("file2.txt")
    combined = d1 + d2
    print(get_hash(combined))


# 4. Sequential Hashing
def sequential_hashing():
    print("\nSequential Hashing")
    d1 = read_file("file1.txt")
    d2 = read_file("file2.txt")

    h1 = get_hash(d1 + "0")
    h2 = get_hash(d2 + h1)

    print(h1)
    print(h2)


# 5. Hierarchical Hashing
def hierarchical_hashing():
    print("\nHierarchical Hashing")
    d1 = read_file("file1.txt")
    d2 = read_file("file2.txt")

    h1 = get_hash(d1)
    h2 = get_hash(d2)

    root = get_hash(h1 + h2)
    print(root)


# Main
independent_hashing()
repeated_hashing()
combined_hashing()
sequential_hashing()
hierarchical_hashing()
```
